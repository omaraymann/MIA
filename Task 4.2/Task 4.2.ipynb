```python
import pandas as pd                                       #to read from excel sheet
from sklearn.linear_model import LinearRegression         
import matplotlib.pyplot as plt                           #to plot the datasheet
from sklearn.model_selection import train_test_split      #to train the model
```

```python
df = pd.read_csv("shrink_ray_dataset.csv")     #read the data sheet values

x = df['knob'].values.reshape(-1,1)            
y = df['Shrinkage'].values.reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #train the model on the knob and shrinkage in the datasheet

model=LinearRegression()
model.fit(x_train,y_train)    

y_predicted = model.predict(x_test)   #the model takes values shrinkage y and predict its knob
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_predicted)

print('Mean Squared Error:', mse)
```

```python
#plots the data sheet points in blue and the red line indicates the linear model
plt.scatter(df['knob'], df['Shrinkage'], label='Data Points')
plt.plot(df['knob'], model.predict(x), color='red')
plt.xlabel('Knob')
plt.ylabel('Shrinkage')
plt.legend()
plt.show()
```

```python
#plots the test values in blue and the linear model of the test value of knobs
plt.scatter(x_test, y_test)
plt.plot(x_test, y_predicted, color='red', linewidth=1)
plt.xlabel('Knob')
plt.ylabel('Shrinkage')
plt.legend()
plt.show()
```

```python
model.fit(y_train,x_train)
desired_shrinkage = 85
predicted_knob = model.predict([[desired_shrinkage]])

print(f"The knob value for a shrinkage percentage of {desired_shrinkage}% is approximately {predicted_knob[0]} MW.")
```
